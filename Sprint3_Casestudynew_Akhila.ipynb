<p><span style="font-size: 15px";>Removing the outliers so we can visualise better.</span></p>

```python
data2 = data[((data['city'] == 'São Paulo') | (data['city'] == 'Rio de Janeiro')) & (data['area'] <800)]
data2
figsize = (10,3)
sb.boxplot(x =data2['area'])
plt.title(f"boxplot with area")
plt.show()

print("50% of the people live in houses of 50 to 200sqft")
```

<p><span style="font-size: 15px";>Lets analyse why people are living in such a small area.</span></p>

```python
bins = [0, 100, 200, 300, 400, 500, 600, 700, 800]
labels = ['0-100', '100-200', '200-300', '300-400', '400-500', '500-600', '600-700', '700-800']
data2['area range'] = pd.cut(data2['area'],bins = bins,labels = labels, right=False)
m = pd.pivot_table(data = data2, index = 'area range', values = {'rent amount (R$)'},aggfunc= ['mean'])
c = pd.pivot_table(data = data2, index = 'area range', values = {'rent amount (R$)'},aggfunc= ['count'])
mval = list(map(int, m.values))
cval = list(map(int, c.values))

fig, ax = plt.subplots(figsize=(8,3))
sb.lineplot(x = m.index,y = mval, ax = ax, marker = 'x',label = 'Average Rent')
sb.lineplot(x = c.index,y = cval, ax = ax, marker = 'o', label = 'Number of houses')

plt.show()

print("As the area of houses increases, the average rent increases and the number of houses decreases.\n So people live in houses in small areas due to fewer houses in large areas and high rents.")
```
